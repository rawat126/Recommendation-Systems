```
%%writefile movie_recommendation.py
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from wordcloud import WordCloud

def main():
    #os.chdir(r'/content/drive/My Drive/Projects/Movie')
    count_data = pd.read_csv(r'/content/drive/My Drive/Projects/Movie Recommendation/count_data.csv')
    correlation_table = pd.read_csv(r'/content/drive/My Drive/Projects/Movie Recommendation/corelation_table.csv')

    all_movies = list(correlation_table.columns)
    all_movies.insert(0,' Select Movies ')

    movie_name = st.selectbox('Select The Last Movie : ', all_movies)
    

    if st.button('Suggest'):
        corr_movies = correlation_table.corrwith(correlation_table[movie_name])
        corr_movies = pd.DataFrame(corr_movies).reset_index()
        corr_movies.columns = ['title','Correlation']

        movies = pd.merge(count_data,corr_movies,on = 'title')
        selected_movies = movies[movies['Counts']>80].sort_values(by = 'Correlation',ascending = False)
        selected_movies = selected_movies.head(20)
    
    else:
        return 0

    st.write('How Do You want to see the Output ?')
    
    if st.button(' Bar Graph '):
        pass

    elif st.button(' Word Cloud '):
        values = dict(zip(selected_movies['title'],selected_movies['Correlation']))
        cloud = WordCloud(background_color = 'white', width = 1500,height = 1000).generate_from_frequencies(values)
        st.image(cloud)

    elif st.button(' Tabular '):
        selected_movies['Correlation'] = selected_movies['Correlation']*100
        st.dataframe(selected_movies[['title','Correlation']])
 

main()
```
